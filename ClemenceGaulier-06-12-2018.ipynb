## Soustraction et repartition par projet sur plus d'un an.

```python
ly = df.launched_year
dy = df.deadline_year
ad = dy - ly
plt.subplots(figsize=(18,7))
ad.value_counts().plot.bar()
```

## Même chose mais pour les mois

```python
lm = df.launched_month

l = lm + ly
dm = df.deadline_month
d = dm + dy
ad2 = d - l
plt.subplots(figsize=(18,7))
ad2.value_counts().plot.bar()
```
